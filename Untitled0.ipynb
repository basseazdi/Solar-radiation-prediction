## **3. Preprocesado**

```python
##########################
x = df_disp.drop(['salida','pwat_ea3_1','uswrf_s1_1','pwat_ea3_1', 'spfh_2m3_1', 'tcolc_e3_1', 'tmin_2m2_1', 'tmin_2m4_1', 'tmp_2m_3_1', 'ulwrf_t1_1', 'uswrf_s1_1'], axis = 1)
y = df_disp['salida']

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.25, shuffle=False)
# Imprime los nombres de las columnas
print(x.columns)
print("dimensiones de X_train: ", X_train.shape)
print("dimensiones de y_train: ", Y_train.shape)
print("dimensiones de X_test: ", X_test.shape)
print("dimensiones de y_test: ", Y_test.shape)
print("\n")
def tipos (x,tipo):
  columns=[col for col in x.columns if x[col].dtype==tipo]
  return columns

def no_tipos (x,tipo):
  columns=[col for col in x.columns if x[col].dtype!=tipo]
  return columns

nm=tipos(df_disp,'float64')
nm.remove('salida')
cc=no_tipos(df_disp,'float64')

# Creamos pipeline de preproceso numérico y categórico
# Primero definimos los parámetros para los imputadores:
imputer_learner_reg=KNeighborsRegressor()
imputer_learner_class= KNeighborsClassifier()
regresor = KNeighborsRegressor()

# preproceso numericos:
num_pipeline = Pipeline([
    ("imputer",IterativeImputer(estimator=imputer_learner_reg)),
    ("scaler",StandardScaler())
])

# preproceso categorico:
cat_pipeline= Pipeline([
    ("encoder",OneHotEncoder(sparse_output=False, handle_unknown = "ignore")),
    ("imputer",IterativeImputer(estimator=imputer_learner_class))
])

# Los juntamos en un preproceso numerico y categorico
preprocessor= ColumnTransformer([
    ("num", num_pipeline, ['apcp_sf2_1', 'apcp_sf3_1', 'apcp_sf4_1', 'apcp_sf5_1', 'dlwrf_s3_1', 'dswrf_s1_1', 'dswrf_s2_1', 'dswrf_s3_1', 'pres_ms1_1', 'pres_ms2_1', 'pres_ms3_1', 'pres_ms4_1', 'pwat_ea2_1', 'pwat_ea4_1', 'spfh_2m1_1', 'spfh_2m2_1', 'spfh_2m4_1', 'spfh_2m5_1', 'tcdc_ea1_1', 'tcdc_ea2_1', 'tcolc_e1_1', 'tcolc_e5_1', 'tmax_2m5_1', 'tmin_2m3_1', 'tmin_2m5_1', 'tmp_2m_2_1','tmp_2m_5_1', 'tmp_sfc1_1', 'tmp_sfc3_1', 'tmp_sfc5_1', 'ulwrf_s2_1', 'ulwrf_s4_1', 'ulwrf_s5_1', 'ulwrf_t5_1', 'uswrf_s3_1', 'uswrf_s4_1', 'uswrf_s5_1']), # Aplicar el pipeline numérico a las columnas numéricas
    ("cat", cat_pipeline, ['apcp_sf1_1', 'dlwrf_s1_1', 'dlwrf_s2_1', 'dlwrf_s4_1', 'dlwrf_s5_1', 'dswrf_s4_1', 'dswrf_s5_1', 'pres_ms5_1', 'pwat_ea1_1', 'pwat_ea5_1', 'tcdc_ea3_1', 'tcdc_ea4_1', 'tcdc_ea5_1', 'tcolc_e2_1', 'tcolc_e4_1', 'tmax_2m1_1', 'tmax_2m2_1', 'tmax_2m3_1', 'tmax_2m4_1', 'tmin_2m1_1', 'tmp_2m_1_1', 'tmp_2m_4_1', 'tmp_sfc2_1', 'tmp_sfc4_1', 'ulwrf_s1_1', 'ulwrf_s3_1', 'ulwrf_t2_1', 'ulwrf_t3_1', 'ulwrf_t4_1', 'uswrf_s2_1']) # Aplicar el pipeline categórico a las columnas categóricas
])

preprocessor.fit(X_train)

# Definición la función de scoring R^2
def r2_scorer(y_true, y_pred):
    return r2_score(y_true, y_pred)

# Definición de la funcion de scoring VE
def EV_scorer(y_true,y_pred):
    return explained_variance_score(y_true, y_pred)

# Envolver el ColumnTransformer en un Pipeline con un nombre
preprocessor= Pipeline([
    ("preprocessor", preprocessor), # Preprocesar los datos con el transformador de columnas
    ('regresor',regresor)
])
# Hiperparámetros
param_grid = {
    'preprocessor__num__imputer': [IterativeImputer(estimator = imputer_learner_reg), SimpleImputer(strategy='mean')],
    'preprocessor__num__scaler': [StandardScaler(),RobustScaler(), MaxAbsScaler()],
    'preprocessor__cat__imputer': [IterativeImputer(estimator = imputer_learner_class), SimpleImputer(strategy='most_frequent')],
    'regresor' : [KNeighborsRegressor()]
    }
tscv = TimeSeriesSplit(n_splits = 3)
# Crear un objeto GridSearchCV para buscar los mejores hiperparámetros
grid_search_R2 = GridSearchCV(preprocessor, param_grid, cv=tscv, scoring=make_scorer(r2_scorer))
grid_search_R2.fit(X_train, Y_train)

grid_search_scale_EV = GridSearchCV(preprocessor,param_grid,cv=tscv,scoring=make_scorer(EV_scorer))
grid_search_scale_EV.fit(X_train, Y_train)

# Gráfica
plt.figure(figsize=(10, 6))

# Plotear datos de entrenamiento
plt.plot(Y_train, label='Datos de Entrenamiento', color='green')

# Plotear datos de prueba
plt.plot(Y_test, label='Datos de Prueba', color='lightblue')

# Línea vertical para indicar el punto de separación
plt.axvline(x=len(Y_train), color='red', linestyle='--', label='Punto de Separación = Día 3285')

# Etiquetas y título
plt.xlabel('Fechas (días)')
plt.ylabel('Radiación Acumulada (W/m^2)')
plt.title('División entre Datos de Entrenamiento y Prueba')
plt.legend()
plt.show()

# Mostramos los mejores parámetros encontrados usando R^2
print("Mejores hiparámetros encontrados:")
print(grid_search_R2.best_params_)
print('La precision del modelo es ', grid_search_R2.best_score_)
# Mostramos los mejores parámetros encontrados usando Variabilidad Explicada
print("Mejores hiparámetros encontrados:")
print(grid_search_scale_EV.best_params_)
print('La precision del modelo es ', grid_search_scale_EV.best_score_)

# Ajusta el modelo con los mejores hiperparámetros en los datos de entrenamiento
best_model= grid_search_R2.best_estimator_
best_model.fit(X_train, Y_train)

# Realiza las predicciones en el conjunto de prueba
y_pred = best_model.predict(X_test)

R_2 = r2_score(Y_test,y_pred) # R^2 externo
print('El R^2 para el mejor modelo es',R_2)

# Ajusta el modelo con los mejores hiperparámetros en los datos de entrenamiento
best_model= grid_search_scale_EV.best_estimator_
best_model.fit(X_train, Y_train)

# Realiza las predicciones en el conjunto de prueba
y_pred = best_model.predict(X_test)

EV = explained_variance_score(Y_test,y_pred) # Precision VE externa
print('La variabilidad explicada para el mejor modelo es',EV)

```
